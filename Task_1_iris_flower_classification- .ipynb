```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
columns = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width', 'Species']
df = pd.read_csv('iris.data', names=columns)
df.head()
```

### Analyzing the Dataset

### Visualizing the Dataset

```python
sns.pairplot(df, hue='Species');
```

```python
sns.scatterplot(x=df['Petal_length'], y=df['Petal_width'], hue=df['Species'], s=100);
```

```python
setosa_df = df[df.Species == 'Iris-setosa']
versicolor_df = df[df.Species == 'Iris-versicolor']
virginica_df = df[df.Species == 'Iris-virginica']

plt.title('Distribution of Sepal Width')
plt.hist([setosa_df['Sepal_width'],versicolor_df['Sepal_width'], virginica_df['Sepal_width']],bins=np.arange(2, 5, 0.25), stacked=True);
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]
print(X)
```

```python
Y_Data = np.array([np.average(X[:, i][Y==j].astype('float32'))
                     for i in range (X.shape[1])             #here X.shape[1]=4
                          for j in (np.unique(Y))])               #np.unique(Y) = three classes in single row
Y_Data_reshaped = Y_Data.reshape(4, 3)
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)
X_axis = np.arange(len(columns)-1)
print(X_axis)
print(Y_Data_reshaped)
print(Y_Data_reshaped[0])
```

```python
width=0.3
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, columns[:4])
plt.xlabel("Features")
plt.ylabel("Value in cm.")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

### Split and Train the Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
print(y_test)
```

## SVR

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
prediction1 = model_svc.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

```python
#Logisitic Regression
from sklearn.linear_model import LogisticRegression

model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python

prediction2 = model_LR.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```
